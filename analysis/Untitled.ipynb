# Basic Statistics

```python
def count(line):
    return [(champs.loc[x[0]]['name'], x[1])for x in dict(line.value_counts()).items()]

def pickrate(line):
    return [(champs.loc[x[0]]['name'], x[1]/len(line)*200)for x in dict(line.value_counts()).items()]
```

# Mapping Death

```python

def setMap(gca):
    gca.set_xlim([0,15000])
    gca.set_ylim([0,15000])
    img = plt.imread("../statics/minimap.png")
    gca.imshow(img,extent=[0, 14900, 0, 14900], interpolation='nearest')
    gca.set_aspect('equal')
    
```

```python
%matplotlib inline

def mapDeath(championId, col0, col1):
    plt.figure(figsize=(16, 16))
    setMap(plt.gca())

    kayleGames = [df[df[col0+"_championId"] == championId], df[df[col1+"_championId"] == championId]]
    for i in range(2):
        column = col0+"_id" if i==0 else col1+"_id"
        color = 'b' if i==0 else 'r'
        for j in range(len(kayleGames[i])):
            m = kayleGames[i].iloc[j]
            try:
                kills = pd.read_csv(f"../supplement/kills/{m.matchId}.csv")
            except FileNotFoundError: continue

            deaths = kills[kills.death==m[column]][kills.timestamp < 60*1000*10]
            plt.scatter(deaths['x'], deaths['y'], color=color)
```
